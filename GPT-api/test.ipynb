```python
import matplotlib.pyplot as plt
import networkx as nx

# 创建一个有向图
G = nx.DiGraph()

# 添加模块节点
modules = [
    "用户管理模块",
    "内容管理模块",
    "搜索模块",
    "统计分析模块",
    "通知模块",
    "系统配置模块"
]

# 添加节点
for module in modules:
    G.add_node(module)

# 定义模块之间的关系（边）
edges = [
    ("用户管理模块", "内容管理模块"),
    ("内容管理模块", "搜索模块"),
    ("内容管理模块", "统计分析模块"),
    ("内容管理模块", "通知模块"),
    ("系统配置模块", "内容管理模块")
]

# 添加边
G.add_edges_from(edges)

# 绘制功能模块划分图
plt.figure(figsize=(10, 6))
pos = nx.spring_layout(G, seed=42)
nx.draw(G, pos, with_labels=True, node_size=3000, node_color='skyblue', font_size=10, font_color='black', font_weight='bold', edge_color='gray', arrows=True)

plt.title('功能模块划分图')
plt.show()
```
