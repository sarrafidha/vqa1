```python
import matplotlib.pyplot as plt
import networkx as nx

# 
G = nx.DiGraph()

# 
modules = [
    "",
    "",
    "",
    "",
    "",
    ""
]

# 
for module in modules:
    G.add_node(module)

# 
edges = [
    ("", ""),
    ("", ""),
    ("", ""),
    ("", ""),
    ("", "")
]

# 
G.add_edges_from(edges)

# 
plt.figure(figsize=(10, 6))
pos = nx.spring_layout(G, seed=42)
nx.draw(G, pos, with_labels=True, node_size=3000, node_color='skyblue', font_size=10, font_color='black', font_weight='bold', edge_color='gray', arrows=True)

plt.title('')
plt.show()
```
